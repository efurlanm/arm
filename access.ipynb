# Access to the JL server running on Termux

This Notebook runs on the PC and configures access to the JL server running on Termux. Assuming Termux and SSH server are running. In my case the `<hostname>` is`a71`.

```python
! ssh -NfT a71
```

The command `ssh -NfT a71` is used to establish a **non-interactive SSH connection** that runs in the **background**.

Start the JL server in Termux:

```python
! ssh -T a71 'jupyter-lab --no-browser --ip=* --IdentityProvider.token='' --notebook-dir=~ >/dev/null 2>&1 &'
```

At this point JL can be accessed on the PC using the address <http://a71:8888/lab?> in the browser.

Mount the file system:

```python
%%bash
sshfs a71:/data/data/com.termux/files/home /mnt/a71 -o "workaround=rename,uid=1000,gid=1000"
```

## Shutdown

```python
%%bash
ssh -T a71 'pkill -f jupyter'
ssh -O exit a71
sudo umount /mnt/a71
```
